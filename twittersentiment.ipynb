```python
#import necessary libraries
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt 
import os
import datetime as dt 
plt.style.use('ggplot')
sentiment_df = pd.read_csv('sentiment_data.csv')
sentiment_df.head()
```

```python
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])
```

```python
sentiment_df = sentiment_df.set_index(['date','symbol'])
sentiment_df
```

```python
#The returns are not so good if we don't consider engagement_ratio instead of twitterComments or twitterLikes
```

```python
sentiment_df['engagement_ratio'] = sentiment_df['twitterComments']/sentiment_df['twitterLikes']
sentiment_df
```

```python
#filter out the best stocks based on engagement 
```

```python
sentiment_df = sentiment_df[(sentiment_df['twitterLikes']>20) & (sentiment_df['twitterComments']>10)] 
sentiment_df 
```

```python
aggregated_df = (sentiment_df.reset_index('symbol').groupby([pd.Grouper(freq='M'),'symbol'])[['engagement_ratio']].mean()) 
aggregated_df['rank'] = (aggregated_df.groupby(level=0)['engagement_ratio'].transform(lambda x: x.rank(ascending=False)))
aggregated_df
```

```python
filtered_df = aggregated_df[aggregated_df['rank']<6]
filtered_df = filtered_df.reset_index(level=1)
filtered_df.index = filtered_df.index+pd.DateOffset(1) 
filtered_df
```

```python
filtered_df = filtered_df.reset_index().set_index(['date','symbol'])
filtered_df
```

```python
#extract the stocks to form portfolios with at the start of each new month 
```

```python
dates = filtered_df.index.get_level_values('date').unique().tolist()
fixed_dates = {}
for d in dates: 
    fixed_dates[d.strftime('%Y-%m-%d')] = filtered_df.xs(d, level=0).index.tolist()
fixed_dates
```

```python
#download fresh stock prices for only selected/shortlisted stocks 
```

```python
stocks_list = sentiment_df.index.get_level_values('symbol').unique().tolist()
prices_df = yf.download(tickers=stocks_list,
                        start='2021-01-01',
                        end = '2023-03-01')
```

```python
#calculate portfolio returns with monthly rebalancing 
```

```python
returns_df = np.log(prices_df['Adj Close']).diff()
```

```python
returns_df = returns_df.dropna()
returns_df
```

```python
portfolio_df = pd.DataFrame()
for start_date in fixed_dates.keys():
    end_date = (pd.to_datetime(start_date)+pd.offsets.MonthEnd()).strftime('%Y-%m-%d')
    print(start_date)
    print(end_date)
    cols = fixed_dates[start_date]
    print(cols)
    temp_df = returns_df[start_date:end_date][cols].mean(axis=1).to_frame('portfolio_return')
    portfolio_df = pd.concat([portfolio_df, temp_df], axis=0)
```

```python
portfolio_df
```

```python
#download NASDAQ/QQQ prices and calculate returns to compare to our strategy
```

```python
qqq_df = yf.download(tickers='QQQ',
                     start='2021-01-01',
                     end='2023-03-01')
qqq_ret = np.log(qqq_df['Adj Close']).diff().to_frame('nasdaq_return')
portfolio_df = portfolio_df.merge(qqq_ret,
                                  left_index=True,
                                  right_index=True)
portfolio_df
```
